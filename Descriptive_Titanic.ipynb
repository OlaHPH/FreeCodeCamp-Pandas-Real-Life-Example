```python
#import libraries
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('white')
```

```python
#load data
df =pd.read_excel('titanic.xlsx')
df.head()
```

```python
survival = ['not survived', 'survived']
age_groups=[0,10,20,30,40,50,60,70,80]
fare_groups=[0,50,100,150,200,250,300,350,400,450,500,550,600]
ports = ['Cherbourg', 'Queenstown', 'Southampton']
df['age_groups'] = pd.cut(df['age'] , bins=age_groups, right=False)
df['fare_groups'] = pd.cut(df['fare'] , bins=fare_groups, right=False)
df
```

###02.Statistical number

```python
#statistical number
df.describe().style.format('{:.2f}')
```

###1. Passenger class distribution

```python
df_class = pd.crosstab(df['pclass'] , 'count')
df_class.columns.name = ''
n=df_class['count'].sum()
df_class['%'] = df_class['count']/n*100
df_class.loc['Total'] = df_class.sum()
df_class.style.format({'count':'{:.0f}','%':'{:.2f}'})
```

```python
df_class.drop('Total')['count'].plot(kind='bar', width=0.3, rot=0)
plt.title("Passenger class distribution in number")
plt.xlabel('Passenger class')
plt.ylabel("Frequency, n=" + str(n))
plt.grid(axis='y')
```

```python
df_class.drop('Total')['%'].plot(kind='bar', width=0.3, rot=0)
plt.title("Passenger class distribution in percentage")
plt.xlabel('Passenger class')
plt.ylabel("Percentage (%), n=" + str(n))
plt.grid(axis='y')
```

### 2. Survival rate

```python
#Passengers' survival rate - frequency table

# survival corresponding 0 - no, 1 - yes:
df_survived = pd.crosstab(df['survived'] , 'count')

# drop col_0
df_survived.columns.name = ''

# replace numbers by names
df_survived.index = survival

# add percentage column
n = df_survived['count'].sum()
df_survived['%'] = df_survived['count']/n*100

# Total row
df_survived.loc['Total'] = df_survived.sum()

df_survived.style.format({'count': '{:.0f}' , '%': '{:.2f}'})
```

```python
#Passengers' survival rate in number - visualization
df_survived.drop('Total')['count'].plot.barh(color=['grey','green'],title='Proportion of passengers survived in number')
plt.xlabel('Frequency, n=' + str(n))
plt.grid(axis='x')
```

```python
#Passengers' survival rate in % - visualization
df_survived.drop('Total')['%'].plot(kind='barh',color=['grey','green'],title='Proportion of passengers survived in percentage')
plt.xlabel('Percentage (%), n=' + str(n))
plt.grid(axis='x')
```

####Did the age distribution of the survivors differ from the age distribution of the deceased?

```python
#Did the age distribution of the survivors differ from the age distribution of the deceased? And so on, and so forth.
df8 = pd.crosstab(df['age_groups'], df['survived'])
df8.columns.name = ''

df8.columns = ['not survived', 'survived']

count_not = df8['not survived'].sum()
count_survive = df8['survived'].sum()

df8['%notsurvived'] = df8['not survived']/count_not*100
df8['%survived'] = df8['survived']/count_survive*100

df8.loc['Total'] = df8.sum()

df8.style.format({'not survived': '{:.0f}' , '%notsurvived': '{:.2f}' , 'survived': '{:.0f}' , '%survived': '{:.2f}'})
```

```python
#Did the age distribution of the survivors differ from the age distribution of the deceased? - Visualization
df8.drop('Total')[['not survived','survived']].plot(kind='bar',rot=0,color=['#a9a9a9','#0000ff'])
plt.xlabel('Frequency, n=' + str(n))
plt.grid(axis='y')
plt.legend(loc=(0.20,-0.20), ncols=20)
```

```python
#Did the age distribution of the survivors differ from the age distribution of the deceased in percentage? - Visualization
df8.drop('Total')[['%notsurvived','%survived']].plot(kind='bar',rot=0,color=['#a9a9a9','#0000ff'])
plt.xlabel('Percentage (%), n=' + str(n))
plt.grid(axis='y')
plt.legend(loc=(0.20,-0.20), ncols=20)
```
